## Question 1(a)

```python
class BayesianClassifier:
    def __init__(self, pi=np.array([0.1, 0.3, 0.25, 0.35]), var=np.array([0.1, 0.1])):
        self.pi = pi
        self.mean = np.array([[-1, -1], [-1, 1], [1,1], [1,1]])
        self.var = var
        self.num_labels = 4
        
    def __prob_x_given_i(self, x, label):
        arr = np.exp(-1 * np.square(x - self.mean[label])/(2*0.1))
        res = np.prod(arr)
        p = (1/(2*np.pi*0.1)) * res
        return p
    
    def __probability_of_error(self, x):
        cost = []
        for i in range(self.num_labels):
            sum = int(0)
            for j in range(self.num_labels):
                if i == j: 
                    continue
                else:
                    sum += self.pi[j] * self.__prob_x_given_i(x, j)
            cost.append(sum)
        self.cost = np.array(cost)
        
    def calculate_probability_of_error(self, x):
        self.__probability_of_error(x)
        
    def get_probability_of_error(self):
        return self.cost
    
    def predict(self, x):
        self.calculate_probability_of_error(x)
        return np.argmin(self.cost)
```

## Question 1(b)

```python
def train_test_split(df):
    X = np.array(df.iloc[:,:-1])
    y = np.array(df.iloc[:,-1:])
    X_train = X[:5000,:]
    X_test = X[5000:, :]
    y_train = y[:5000,:]
    y_test = y[5000:, :]
    return X_train, X_test, y_train, y_test
```

```python
# calculates estimate of prior probabilities for different classes
def estimate_pi(y_train):
    counts = [int(0)]*4
    for i in range(y_train.shape[0]):
        counts[int(y_train[i])] += 1
    pi_cap = np.array(counts)
    pi_cap = pi_cap/y_train.shape[0]
    return pi_cap
```

```python
def estimate_sigma(x_train):
    mean = np.mean(x_train, axis=0)
    sigma = np.sum(np.square(x_train - mean), axis=0)/x_train.shape[0]
    return sigma
```

```python
df = pd.read_csv("Dataset1_Assignment1.csv")
X_train, X_test, y_train, y_test = train_test_split(df)
```

```python
# Question 1(b) part 1
bc = BayesianClassifier()
i = int(1)
for x in X_test:
    print(f"Test case {i}: Predicted Label={bc.predict(x)}, Probability of Error of all classes={bc.get_probability_of_error()}")
    i+=1
```

```python
# Question 1(b) part 2
pi_estimator = estimate_pi(y_train)
sigma_estimate = estimate_sigma(X_train)
bc = BayesianClassifier(pi_estimator, sigma_estimate)
print(f"pi_cap = {pi_estimator}")
print(f"sigma_cap = {sigma_estimate}")
```

```python
i = int(1)
for x in X_test:
    print(f"Test case {i}: Predicted Label={bc.predict(x)}, Probability of Error of all classes={bc.get_probability_of_error()}")
    i+=1
```
